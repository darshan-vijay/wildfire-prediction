```python
# Instrument : MODIS_SP

#url = https://firms.modaps.eosdis.nasa.gov/api/area/csv/' + MAP_KEY + '/MODIS_SP/112.467,-55.05,168,-9.133/1/'

data = pd.DataFrame()

from datetime import datetime, timedelta

start = datetime(2000, 11, 1)
end = datetime(2012, 1, 19)

# Loop through each day
current_date = start
while current_date <= end:
    
    dt = current_date.strftime('%Y-%m-%d')
    
    area_url = f'https://firms.modaps.eosdis.nasa.gov/api/area/csv/{MAP_KEY}/MODIS_SP/112.467,-55.05,168,-9.133/1/{dt}'
    
    try:
        df_area = pd.read_csv(area_url)
        data = pd.concat([data, df_area], ignore_index=True)
    except Exception as e:
        print(f"Error fetching data for {dt}: {e}")
    
    # Move to the next date
    current_date += timedelta(days=1)

```

```python
def get_state(lat, lon):
    if -37.505 <= lat <= -28.157 and 140.999 <= lon <= 153.639:
        return 'New South Wales'
    elif -39.159 <= lat <= -33.980 and 140.961 <= lon <= 150.066:
        return 'Victoria'
    elif -29.177 <= lat <= -9.142 and 138.000 <= lon <= 153.552:
        return 'Queensland'
    elif -35.137 <= lat <= -13.689 and 112.921 <= lon <= 129.000:
        return 'Western Australia'
    elif -38.065 <= lat <= -25.996 and 129.000 <= lon <= 141.000:
        return 'South Australia'
    elif -43.742 <= lat <= -39.439 and 143.818 <= lon <= 148.479:
        return 'Tasmania'
    elif -26.000 <= lat <= -10.982 and 129.000 <= lon <= 138.000:
        return 'Northern Territory'
    elif -35.920 <= lat <= -35.124 and 148.763 <= lon <= 149.399:
        return 'Australian Capital Territory'
    else:
        return 'Unknown'
```

```python
df = pd.read_csv('Datasets/fire_data.csv')
```

```python
df['state'] = df.apply(lambda row: get_state(row['latitude'], row['longitude']), axis=1)
```

```python
df['state'].value_counts()
```

```python
df['acq_date'] = pd.to_datetime(df['acq_date'])
```

```python
df['confidence'] = pd.to_numeric(df['confidence'], errors='coerce')

queensland_filtered_data = df.loc[
    (df['state'] == 'Queensland') & 
    (df['acq_date'].dt.year >= 2022) & 
    (df['confidence'] > 33)
]
```
